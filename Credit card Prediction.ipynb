# **Data Visualization**

```python
# Histogram of the 'Age' column
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of 'Income' by 'Gender'
plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Income', data=df)
plt.title('Income Distribution by Gender')
plt.show()
```

```python
plt.figure(figsize=(6,6))
colors = ['#ff9999', '#66b3ff']
explode = (0, 0.1)
plt.pie(x=df['Gender'].value_counts().values, labels=['Female', 'Male'],
        autopct='%.1f%%', startangle=90, explode=explode, colors=colors, shadow=True)
plt.legend(title="Gender", loc="upper right")
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,6))
colors = ['#ff9999', '#66ff99']
explode = (0, 0.1)

# Count 'Yes' and 'No' values in the 'Has a car' column
yes_count = df['Has a car'].value_counts()['Y']
no_count = df['Has a car'].value_counts()['N']

# Create the pie chart
plt.pie(x=[yes_count, no_count], labels=['Yes', 'No'],
        autopct=lambda p: '{:.1f}%\n({:.0f})'.format(p, p * sum([yes_count, no_count]) / 100),
        startangle=90, explode=explode, colors=colors, shadow=True)

plt.legend(title="Has a car", loc="upper right")
plt.title("Has a car")
plt.show()
```

```python
num = df['Children count'].value_counts().values
label = df['Children count'].value_counts().index
total = len(df)
plt.figure(figsize=(10,10))
ax = sns.barplot(x=label, y=num, palette='rocket')
ax.bar_label(ax.containers[0])
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height / total:.1%}',
                (p.get_x() + p.get_width() / 2., height/2),
                ha='center', va='center', fontsize=8, color='#66b3ff', fontweight='bold')

ax.set_ylabel("Count")
ax.set_xlabel("Number of Children")

# Add a title
ax.set_title("Distribution of Number of Children")

plt.show()
```

```python
plt.figure(figsize=(10, 6))

ax = sns.histplot(data=df[df['Income'] <= 600000], x='Income', bins=30,
                  color='skyblue', edgecolor='black', alpha=0.7)
ax.set_xlabel('Income')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Income')
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.tick_params(axis='both', which='major', labelsize=10)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.barplot(
            y=df['Job title'].value_counts().index,
            x=df['Job title'].value_counts().values,
            palette='rocket')

ax = plt.gca()
for i, v in enumerate(df['Job title'].value_counts().values):
    ax.text(v + 3, i, str(v), color='black', va='center')

plt.xlabel('Count')
plt.ylabel('Occupation Type')
plt.title('Distribution of Occupation Types')

plt.show()
```

# **MODEL**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd

# Load train and test datasets from separate CSV files
train_df = pd.read_csv("train_data.csv")
test_df = pd.read_csv("test_data.csv")

# Splitting the train dataset into features (X_train) and target variable (y_train)
X_train = train_df.drop('Is high risk', axis=1)  # Features
y_train = train_df['Is high risk']  # Target variable

# Splitting the test dataset into features (X_test) and target variable (y_test)
X_test = test_df.drop('Is high risk', axis=1)  # Features
y_test = test_df['Is high risk']  # Target variable

# One-hot encode categorical variables
X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)

# Initialize and train various machine learning models
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier()
}

for name, model in models.items():
    model.fit(X_train_encoded, y_train)
    y_pred = model.predict(X_test_encoded)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Model: {name}")
    print(f"Accuracy: {accuracy:.2f}")
    print(classification_report(y_test, y_pred))
    print("\n")
```
